```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import PIL

img = PIL.Image.open('./Playboi-Carti.jpg')
img_arr = np.array(img)

r = []
g = []
b = []
for i in range(len(img_arr)):
    for j in range(len(img_arr[i])):
        pixel = img_arr[i][j]
        r.append(pixel[0])
        g.append(pixel[1])
        b.append(pixel[2])

ax = plt.axes(projection='3d')
ax.scatter3D(r, g, b)
```
